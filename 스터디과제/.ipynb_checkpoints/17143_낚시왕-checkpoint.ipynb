```python
r,c,N = map(int,input().split())
matrix = []
sharks = {}
for i in range(N):
    matrix.append(list(map(int,input().split())))
    
for i in range(N):
    sharks[(matrix[i][0],matrix[i][1])] = (matrix[i][2],matrix[i][3],matrix[i][4])
    
temp_list_r_1 = [i for i in range(r,1,-1)] + [i for i in range(1,r)]
temp_list_r_2 = [i for i in range(1,r+1)] + [i for i in range(r-1,1,-1)]
temp_list_c_1 = temp_list = [i for i in range(1,c+1)] + [i for i in range(c-1,1,-1)]
temp_list_c_2 = temp_list = [i for i in range(c,1,-1)] + [i for i in range(1,c)]

def move(row,temp_list_r_1,temp_list_r_2,temp_list_c_1,temp_list_c_2,r,c):

    if row[3] == 1: ## 위

        one = (r - row[0] + row[2]) % len(temp_list_r_1)
        direction = (r - row[0]+ row[2]) // (r-1)
        if direction % 2 == 0:
            pass
        else:
            row[3] = 2 
        row[0] = temp_list_r_1[one]
        return row

    elif row[3] == 2: ## 아래

        temp_list
        one = (row[0] + row[2] -1) % len(temp_list_r_2)
        direction = (row[0]+ row[2] -1) // (r-1)
        if direction % 2 == 0:
            pass
        else:
            row[3] = 1
        row[0] = temp_list_r_2[one]
        return row

    elif row[3] == 3: ## 오른쪽

        temp_list
        one = (row[1] + row[2] -1) % len(temp_list_c_1)
        direction = (row[1]+ row[2] -1) // (c-1)
        if direction % 2 == 0:
            pass
        else:
            row[3] = 4

        row[1] = temp_list_c_1[one]
        return row


    ## 왼쪽
    else:
        
        one = (c - row[1] + row[2]) % len(temp_list_c_2)
        direction = (c - row[1]+ row[2]) // (c-1)
        if direction % 2 == 0:
            pass
        else:
            row[3] = 3

        row[1] = temp_list_c_2[one]
        return row
    
answer = 0
for i in range(1,c+1):
    for j in range(1,r+1):
        if (j,i) in sharks:
            answer += sharks[(j,i)][2]
            del sharks[(j,i)]
            break
    

    new_sharks = {}
    
    for shark in sharks:
        nr,nc = shark[0],shark[1]
        s,d,z = sharks[(nr,nc)]
        new_row = move([nr,nc,s,d,z],temp_list_r_1,temp_list_r_2,temp_list_c_1,temp_list_c_2,r,c)
        
        if (new_row[0],new_row[1]) not in new_sharks:
            new_sharks[(new_row[0],new_row[1])] = (new_row[2],new_row[3],new_row[4])
        else:
            if new_sharks[(new_row[0],new_row[1])][2] < z:
                new_sharks[(new_row[0],new_row[1])] = (new_row[2],new_row[3],new_row[4])
    
    sharks = new_sharks
    
print(answer)
        
    
```
