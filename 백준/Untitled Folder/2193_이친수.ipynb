```python
n = int(input())
max_value = 90

matrix = {}
matrix[(1,1)] = 1
matrix[(1,0)] = 0

for i in range(2,max_value+1):
    matrix[(i,1)] = matrix[(i-1,0)]
    matrix[(i,0)] = matrix[(i-1,0)] + matrix[(i-1,1)]
    
answer = 0
    
for i in range(2):
    answer += matrix[(n,i)]
    
print(answer)
```
