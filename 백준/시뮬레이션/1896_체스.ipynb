```python
a,b = map(int,input().split())

map_dict = {}

for i in range(a):
    for j in range(b):
        map_dict[(i+1,j+1)] = 1
        
Q_list = list(map(int,input().split()))
K_list = list(map(int,input().split()))
P_list = list(map(int,input().split()))

for i in range(Q_list[0]):
    map_dict[(Q_list[2*i + 1],Q_list[2*i + 2])] = 'Q'
    
for i in range(K_list[0]):
    map_dict[(K_list[2*i + 1],K_list[2*i + 2])] = 'K'
    
for i in range(P_list[0]):
    map_dict[(P_list[2*i + 1],P_list[2*i + 2])] = 'P'
    
def knight(x,y):
    if ((x-1)>0)&((y-2)>0):
        if type(map_dict[(x-1,y-2)]) != str:
            map_dict[(x-1,y-2)] = 0
    if ((x-2)>0)&((y-1)>0):
        if type(map_dict[(x-2,y-1)]) != str:
            map_dict[(x-2,y-1)] = 0
    if ((x-2)>0)&((y+1)<=b):
        if type(map_dict[(x-2,y+1)]) != str:
            map_dict[(x-2,y+1)] = 0
    if ((x-1)>0)&((y+2)<=b):
        if type(map_dict[(x-1,y+2)]) != str:
            map_dict[(x-1,y+2)] = 0
    if ((x+1)<=a)&((y-2)>0):
        if type(map_dict[(x+1,y-2)]) != str:
            map_dict[(x+1,y-2)] = 0
    if ((x+2)<=a)&((y-1)>0):
        if type(map_dict[(x+2,y-1)]) != str:
            map_dict[(x+2,y-1)] = 0
    if ((x+2)<=a)&((y+1)<=b):
        if type(map_dict[(x+2,y+1)]) != str:
            map_dict[(x+2,y+1)] = 0
    if ((x+1)<=a)&((y+2)<=b):
        if type(map_dict[(x+1,y+2)]) != str:
            map_dict[(x+1,y+2)] = 0
        
def queen_x1(x,y):
    while True:
        x -= 1

        if  (x < 1) or (type(map_dict[(x,y)]) == str):
            break
            
        map_dict[(x,y)] = 0
            
def queen_x2(x,y):
    while True:
        x += 1

        if  (x > a) or (type(map_dict[(x,y)]) == str):
            break
            
        map_dict[(x,y)] = 0
            
def queen_y1(x,y):
    while True:
        y-=1

        if (y < 1) or (type(map_dict[(x,y)]) == str):
            break
            
        map_dict[(x,y)] = 0
            
def queen_y2(x,y):
    while True:
        y += 1

        if (y > b) or (type(map_dict[(x,y)]) == str):
            break
            
        map_dict[(x,y)] = 0
        
def queen_z1(x,y):
    while True:
        y-=1
        x-=1

        if  (y < 1) or (x < 1) or (type(map_dict[(x,y)]) == str):
            break
            
        map_dict[(x,y)] = 0
        
def queen_z2(x,y):
    while True:
        x-=1
        y+=1

        if (y > b) or (x < 1) or (type(map_dict[(x,y)]) == str):
            break
            
        map_dict[(x,y)] = 0
        
        
def queen_z3(x,y):
    while True:
        x += 1
        y -= 1

        if (y < 1) or (x > a) or (type(map_dict[(x,y)]) == str):
            break
        
        map_dict[(x,y)] = 0
        
def queen_z4(x,y):
    while True:
        x += 1
        y += 1

        if (x > a) or (y > b) or (type(map_dict[(x,y)]) == str):
            break
            
        map_dict[(x,y)] = 0
        
        
def all_direction(x,y,kind):
    if kind == 'Q':
        queen_x1(x,y)
        queen_x2(x,y)
        queen_y1(x,y)
        queen_y2(x,y)
        queen_z1(x,y)
        queen_z2(x,y)
        queen_z3(x,y)
        queen_z4(x,y)
        
    if kind == 'K':
        knight(x,y)       

for i in range(Q_list[0]):
    all_direction(Q_list[2*i + 1],Q_list[2*i + 2],'Q') 

    
for i in range(K_list[0]):
    all_direction(K_list[2*i + 1],K_list[2*i + 2], 'K')

    
for i in range(P_list[0]):
    all_direction(P_list[2*i + 1],P_list[2*i + 2], 'P')

    
for i in range(Q_list[0]):
    map_dict[(Q_list[2*i + 1],Q_list[2*i + 2])] = 0
    
for i in range(K_list[0]):
    map_dict[(K_list[2*i + 1],K_list[2*i + 2])] = 0
    
for i in range(P_list[0]):
    map_dict[(P_list[2*i + 1],P_list[2*i + 2])] = 0
    
print(sum(map_dict.values()))
```
