```python
def do(maps,x,y,direction):

    if [x,y,direction] in visited:
        print(0,0,end = ' ')
        return
    
    if (x == 0) or (x == (len(maps)-1)):
        print(x,y,end = ' ')
        return
    
    if (y == 0) or (y == (len(maps)-1)):
        print(x,y,end = ' ')
        return
    
    if direction == 'up':
        if maps[x][y] == '*':
            visited.append([x,y,direction])
            do(maps,x,y+1,'right')
        else:
            visited.append([x,y,direction])
            do(maps,x-1,y,'up')
            
        
    elif direction == 'down':
        if maps[x][y] == '*':
            visited.append([x,y,direction])
            do(maps,x,y-1,'left')

        else:
            visited.append([x,y,direction])
            do(maps,x+1,y,direction)

    elif direction == 'left':
        if maps[x][y] == '*':
            visited.append([x,y,direction])
            do(maps,x-1,y,'up')
            
        else:
            visited.append([x,y,direction])
            do(maps,x,y-1,direction)
            
        
    elif direction == 'right':
        if maps[x][y] == '*':
            visited.append([x,y,direction])
            do(maps,x+1,y,'down')
            
        else:
            visited.append([x,y,direction])
            do(maps,x,y+1,direction)
            
            
def start_direction(a,x,y):
    if mirror[0] == a+1:
        return x-1,y,'up'
    
    elif mirror[1] == 0:
        return x,y+1,'right'
    
    elif mirror[1] == a+1:
        return x,y-1,'left'
    
    elif mirror[0] == 0:
        return x+1,y,'down'

n= int(input())
for i in range(n):
    matrix = []

    a,b = map(int,input().split())
    for j in range(b):
        matrix.append(list(map(int,input().split())))
        
        
    mirror = list(map(int,input().split()))
    visited = []
    maps = [[0]*(a+2) for _ in range(a+2)]
    start_x,start_y,direction = start_direction(a,mirror[0],mirror[1])
    
    for k in range(len(matrix)):
        x1 = matrix[k][0]
        y1 = matrix[k][1]
        maps[x1][y1] = '*'
        
    do(maps,start_x,start_y,direction)
    
```
