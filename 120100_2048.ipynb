```python
from copy import deepcopy


def line(row_list):
    copy = row_list[:]
    for j in range(len(copy)-1):
        for i in range(len(copy)-1):
            if copy[i] == 0:
                copy[i] = copy[i+1]
                copy[i+1] =  0
    temp_list = []  
    
    idx = 0
    while idx < len(copy) :      
        if copy[idx] == 0:
            idx += 1
            
        elif idx == len(copy)-1:
            temp_list.append(copy[idx])
            idx +=1

        elif copy[idx] == copy[idx+1]:
            temp_list.append(copy[idx]*2)
            idx += 2

        else:
            temp_list.append(copy[idx])
            idx += 1
    
    for i in range(len(copy)-len(temp_list)):
        temp_list.append(0)
        
    
    return temp_list


def matrix_op(matrix):
    temp = []
    for i in range(n):
        temp.append(line(matrix[i]))
        
    return temp

def matrix_max(matrix):
    a = 0
    for i in range(n):
        for j in range(n):
            if matrix[i][j] > a:
                a = matrix[i][j]
    return a

def rotate0(matrix):
    temp_matrix = []
    temp_matrix = matrix_op(matrix)
    return temp_matrix



def rotate1(matrix): ## right
    new_matrix = []
    temp_matrix = []
    
    for i in range(n):
        temp_list = []
        for k in range(n-1,-1,-1):
            temp_list.append(matrix[i][k])
        new_matrix.append(temp_list)
        
    temp_matrix = matrix_op(new_matrix)
        
    temp = []
    for i in range(n):
        temp_list = []
        for j in range(n-1,-1,-1):
            temp_list.append(temp_matrix[i][j])
        temp.append(temp_list)
            
    return temp

def rotate2(matrix): ## up
    new_matrix = []
    temp_matrix = []
    
    for i in range(n):
        new_matrix.append([x[i] for x in matrix])
        
    temp_matrix = matrix_op(new_matrix)
    
    temp = []    
    for i in range(n):
        temp.append([x[i] for x in temp_matrix])
        
    return temp
    
    

def rotate3(matrix): ## down
    new_matrix = []
    temp_matrix = []
    
    for i in range(n):
        temp_list = []
        for k in range(n-1,-1,-1):
            temp_list.append(matrix[k][i])
            
        new_matrix.append(temp_list)
        
    temp_matrix = matrix_op(new_matrix)
    
    temp = []
    
    for i in range(n-1,-1,-1):
        temp.append([x[i] for x in temp_matrix])
        
    return temp
    
        
    
def move(rotate,matrix,count):
    global max_value
    new_matrix = []
    
    if rotate == 0:
        new_matrix = rotate0(matrix)
        
    elif rotate == 1:
        new_matrix = rotate1(matrix)
        
    elif rotate == 2:
        new_matrix = rotate2(matrix)
        
    elif rotate == 3:
        new_matrix = rotate3(matrix)
    
    if count == 5:
        max_value = max(max_value,matrix_max(new_matrix))
        return
    
    
    

    move(0,new_matrix,count+1)
    move(1,new_matrix,count+1)
    move(2,new_matrix,count+1)
    move(3,new_matrix,count+1)

    
n = int(input())

matrix = []
for i in range(n):
    row = list(map(int,input().split()))
    matrix.append(row)

max_value = 0
    
original_matrix = deepcopy(matrix)    
for i in range(4):
    move(i,original_matrix,1)
                    
print(max_value) 
    
    
```
