```python
a,b = map(int,input().split())

matrix = []
MZ = 'M_index'

for i in range(a):
    matrix.append(list(input()))
    
# M index
M_index = []
for i in range(a):
    for j in range(b):
        if matrix[i][j] == 'M':
            M_index.append(i)
            M_index.append(j)
            
# Z index
Z_index = []
for i in range(a):
    for j in range(b):
        if matrix[i][j] == 'Z':
            Z_index.append(i)
            Z_index.append(j)
            
answer = []
arrive = []

def move(x,y,before): ## 양방향
    
    if x > a-1:
        return
    if y > b-1:
        return

    if matrix[x][y] == '|':
        if before == [x-1,y]:
            move(x+1,y,[x,y])
        elif before == [x+1,y]:
            move(x-1,y,[x,y])
        else:
            return
            
    if matrix[x][y] == '-':
        if before == [x,y-1]:
            move(x,y+1,[x,y])
        elif before == [x,y+1]:
            move(x,y-1,[x,y])
        else:
            return
            
    if matrix[x][y] == '+':
        if before == [x-1,y]: ## 수직 down
            move(x+1,y,[x,y])
        elif before == [x+1,y]: ## 수직 up
            move(x-1,y,[x,y])
        elif before == [x,y-1]: ## 수평 right
            move(x,y+1,[x,y])
        elif before == [x,y+1]: ## 수평 left
            move(x,y-1,[x,y])
        else:
            return

        
    if matrix[x][y] == '1':
        if before == [x,y+1]:
            move(x+1,y,[x,y])
        elif before == [x+1,y]:
            move(x,y+1,[x,y])
        else:
            return
    if matrix[x][y] == '2':
        if before == [x-1,y]:
            move(x,y+1,[x,y])
        elif before == [x,y+1]:
            move(x-1,y,[x,y])
        else:
            return
    if matrix[x][y] == '3':
        if before == [x,y-1]:
            move(x-1,y,[x,y])
        elif before == [x-1,y]:
            move(x,y-1,[x,y])
        else:
            return
    if matrix[x][y] == '4':
        if before == [x,y-1]:
            move(x+1,y,[x,y])
        elif before == [x+1,y]:
            move(x,y-1,[x,y])
        else:
            return
        
    if matrix[x][y] == ".":
        answer.append(x)
        answer.append(y)
        
    if matrix[x][y] == 'Z':
        arrive.append('1')
        
    if matrix[x][y] == 'M':
        arrive.append('1')

        
        

    
def findBlank(x,y):
    if x != 0:
        if matrix[x-1][y] != ".": ## up
            move(x-1,y,[x,y])

    if x != a-1:
        if matrix[x+1][y] != ".": ## down
            move(x+1,y,[x,y])

    if y != 0:
        if matrix[x][y-1] != ".":  ## left
            move(x,y-1,[x,y])

    if y != b-1:
        if matrix[x][y+1] != ".": ## right
            move(x,y+1,[x,y])
            

        
            

    
            
findBlank(M_index[0],M_index[1]) ## 예외처리

if len(answer) == 0:
    findBlank(Z_index[0],Z_index[1])
    MZ = 'Z_index'



answer_list = ['|','-','+','1','2','3','4']
arrive = []
final_answer = answer[:]

def check(MZ):
    if MZ == 'M_index':

        for i in range(len(answer_list)):
            matrix[answer[0]][answer[1]] = answer_list[i]
            findBlank(M_index[0],M_index[1])
            if len(arrive) != 0:
                final_answer.append(answer_list[i])
                break
    else: 

        for i in range(len(answer_list)):
            matrix[answer[0]][answer[1]] = answer_list[i]
            findBlank(Z_index[0],Z_index[1])
            if len(arrive) != 0:
                final_answer.append(answer_list[i])
                break        

check(MZ)

final_answer[0] += 1
final_answer[1] += 1

for i in range(len(final_answer)):
    print(final_answer[i],end = ' ')
```
