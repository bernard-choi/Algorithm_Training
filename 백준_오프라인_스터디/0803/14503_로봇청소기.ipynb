```python
## 실제 좌표값을 넣는다고 가정
answer = 1


def move(x,y,d,visited):

    print(x,y,d)
    print(visited)
    
    global answer
    print(answer)

    
    if d == 0: ## 위
        
        
        if ((matrix[x][y-1] == 1) or ([x,y-1] in visited)) & ((matrix[x-1][y] == 1) or ([x-1,y] in visited )) & ((matrix[x][y+1] == 1) or ([x,y+1] in visited)) &((matrix[x+1][y] == 1) or ([x+1,y] in visited)):
            if matrix[x+1][y] == 1: 
 ## 네 방향 모두 청소가 이미 되어있거나 벽이면서, 뒤쪽 방향이 벽이라 후진도 할 수 없는 경우에는 작동을 멈춘다.
                return
            else:
                move(x+1,y,d,visited) ## 네 방향 모두 청소가 이미 되어있거나 벽인 경우에는, 바라보는 방향을 유지한 채로 한 칸 후진을 하고 2번으로 돌아간다.

        elif (matrix[x][y-1] == 1) or ([x,y-1] in visited): ## 왼쪽 방향에 청소할 공간이 없다면, 그 방향으로 회전하고 2번으로 돌아간다.
            d=3
            move(x,y,d,visited)
            
            
        else: # 왼쪽 방향에 아직 청소하지 않은 공간이 존재한다면, 그 방향으로 회전한 다음 한 칸을 전진하고 1번부터 진행한다.
            d=3
            answer += 1

            move(x,y-1,d,visited+[[x,y-1]])
            
                
    if d == 1: ## 오른쪽

        
        if ((matrix[x][y-1] == 1) or ([x,y-1] in visited)) & ((matrix[x-1][y] == 1) or ([x-1,y] in visited)) & ((matrix[x][y+1] == 1) or ([x,y+1] in visited)) &((matrix[x+1][y] == 1) or ([x+1,y] in visited)):
            if matrix[x][y-1] == 1: ## 네 방향 모두 청소가 이미 되어있거나 벽이면서, 뒤쪽 방향이 벽이라 후진도 할 수 없는 경우에는 작동을 멈춘다.
                return
            else:
                move(x,y-1,d,visited) ## 네 방향 모두 청소가 이미 되어있거나 벽인 경우에는, 바라보는 방향을 유지한 채로 한 칸 후진을 하고 2번으로 돌아간다.

        elif (matrix[x-1][y] == 1) or ([x-1,y] in visited): ## 왼쪽 방향에 청소할 공간이 없다면, 그 방향으로 회전하고 2번으로 돌아간다.
            d=0
            move(x,y,d,visited)
            
        else: # 왼쪽 방향에 아직 청소하지 않은 공간이 존재한다면, 그 방향으로 회전한 다음 한 칸을 전진하고 1번부터 진행한다.
            d=0
            answer += 1
            move(x-1,y,d,visited+[[x-1,y]])
            
                
    if d == 2: ## 아래
        
        
        if ((matrix[x][y-1] == 1) or ([x,y-1] in visited)) & ((matrix[x-1][y] == 1) or ([x-1,y] in visited )) & ((matrix[x][y+1] == 1) or ([x,y+1] in visited)) &((matrix[x+1][y] == 1) or ([x+1,y] in visited)):
            if matrix[x-1][y] == 1: ## 네 방향 모두 청소가 이미 되어있거나 벽이면서, 뒤쪽 방향이 벽이라 후진도 할 수 없는 경우에는 작동을 멈춘다.
                return
            else:
                move(x-1,y,d,visited) ## 네 방향 모두 청소가 이미 되어있거나 벽인 경우에는, 바라보는 방향을 유지한 채로 한 칸 후진을 하고 2번으로 돌아간다.

        elif (matrix[x][y+1] == 1) or ([x,y+1] in visited): ## 왼쪽 방향에 청소할 공간이 없다면, 그 방향으로 회전하고 2번으로 돌아간다.
            d=1
            move(x,y,d,visited)
            
        else: # 왼쪽 방향에 아직 청소하지 않은 공간이 존재한다면, 그 방향으로 회전한 다음 한 칸을 전진하고 1번부터 진행한다.
            d=1
            answer += 1
            move(x,y+1,d,visited+[[x,y+1]])
            

    if d == 3: ## 왼쪽
        
        
        if ((matrix[x][y-1] == 1) or ([x,y-1] in visited)) & ((matrix[x-1][y] == 1) or ([x-1,y] in visited)) & ((matrix[x][y+1] == 1) or ([x,y+1] in visited)) &((matrix[x+1][y] == 1) or ([x+1,y] in visited)):
            if matrix[x][y+1] == 1: ## 네 방향 모두 청소가 이미 되어있거나 벽이면서, 뒤쪽 방향이 벽이라 후진도 할 수 없는 경우에는 작동을 멈춘다.
                return
            else:
                move(x,y+1,d,visited) ## 네 방향 모두 청소가 이미 되어있거나 벽인 경우에는, 바라보는 방향을 유지한 채로 한 칸 후진을 하고 2번으로 돌아간다.
 
        
        elif (matrix[x+1][y] == 1) or ([x+1,y] in visited): ## 왼쪽 방향에 청소할 공간이 없다면, 그 방향으로 회전하고 2번으로 돌아간다.
            d=2
            move(x,y,d,visited)
            
        else: # 왼쪽 방향에 아직 청소하지 않은 공간이 존재한다면, 그 방향으로 회전한 다음 한 칸을 전진하고 1번부터 진행한다.
            d=2
            answer += 1
            move(x+1,y,d,visited+[[x+1,y]])
            
move(x,y,d,visited = [[x,y]])
            

print(answer)
```
